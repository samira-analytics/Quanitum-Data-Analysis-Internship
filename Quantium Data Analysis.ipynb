```python
import pandas as pd
import numpy as np
import re
import matplotlib.pyplot as plt
import seaborn as sns

# --- 1. Load and Inspect the Raw Transaction Data ---

# Load the data
transaction_data = pd.read_excel(r"C:\Users\patel\Downloads\QVI_transaction_data.xlsx")
purchase_behaviour = pd.read_csv(r"C:\Users\patel\Downloads\QVI_purchase_behaviour.csv")
```

```python
print("### Transaction Data Summary")
print(transaction_data.info())
print(transaction_data.head())
```

```python

print("\n### Purchase Behaviour Summary")
print(purchase_behaviour.info())
print(purchase_behaviour.head())
```

```python
print("\n### Null Values Check")
print("Transaction Data:\n", transaction_data.isnull().sum())
print("Purchase Behaviour:\n", purchase_behaviour.isnull().sum())
```

```python
print("\n### Duplicate Rows Check")
print("Transaction Data Duplicates:", transaction_data.duplicated().sum())
print("Purchase Behaviour Duplicates:", purchase_behaviour.duplicated().sum())
```

```python
# --- 2. Data Cleaning, Outlier Detection, and Feature Engineering ---
# Convert DATE (Excel format) to datetime
transaction_data["DATE"] = pd.to_datetime(transaction_data["DATE"], unit="D", origin="1899-12-30")
```

```python
# Extract Pack Size
transaction_data["PACK_SIZE"] = transaction_data["PROD_NAME"].str.extract(r"(\\d+)").astype(float)
```

```python
# Filter for Chips only (remove salsa)
transaction_data = transaction_data[~transaction_data["PROD_NAME"].str.lower().str.contains("salsa")]
```

```python
# Extract Brand Name
transaction_data["BRAND"] = transaction_data["PROD_NAME"].str.split().str[0]
```

```python
# Clean brand names
brand_map = {
    "Red": "RRD",
    "Smith": "Smiths",
    "Snbts": "Sunbites",
    "Infzns": "Infuzions",
    "WW": "Woolworths",
    "Dorito": "Doritos",
    "Grain": "Gransaves" # Assuming 'Grain' refers to Gransaves based on common chip brands
}
transaction_data["BRAND"] = transaction_data["BRAND"].replace(brand_map)
```

```python
# Outlier Detection for PROD_QTY
print("\n### Summary Statistics for PROD_QTY (before outlier removal)")
print(transaction_data["PROD_QTY"].describe())
```

```python
# Remove outliers (customer who bought 200 packets twice)
transaction_data = transaction_data[transaction_data["PROD_QTY"] < 100]
```

```python
# Merge dataframes
data = pd.merge(transaction_data, purchase_behaviour, on="LYLTY_CARD_NBR", how="left")
```

```python
print("\n### Missing values after merge")
print(data.isnull().sum())
```

```python
# --- 3. Customer Segment Analysis and Metric Derivation ---
# Total Sales by LIFESTAGE and PREMIUM_CUSTOMER
sales_by_segment = data.groupby(["LIFESTAGE", "PREMIUM_CUSTOMER"])["TOT_SALES"].sum().reset_index()
sales_by_segment.rename(columns={
    "TOT_SALES": "Total Sales"
}, inplace=True)
```

```python
# Number of Customers by segment
customers_by_segment = data.groupby(["LIFESTAGE", "PREMIUM_CUSTOMER"])["LYLTY_CARD_NBR"].nunique().reset_index()
customers_by_segment.rename(columns={
    "LYLTY_CARD_NBR": "Number of Customers"
}, inplace=True)
```

```python
# Average number of units per customer by segment
units_per_customer = data.groupby(["LIFESTAGE", "PREMIUM_CUSTOMER"])["PROD_QTY"].sum() / \
                     data.groupby(["LIFESTAGE", "PREMIUM_CUSTOMER"])["LYLTY_CARD_NBR"].nunique()
units_per_customer = units_per_customer.reset_index(name="Avg Units per Customer")
```

```python
# Average price per unit by segment
price_per_unit = data.groupby(["LIFESTAGE", "PREMIUM_CUSTOMER"])["TOT_SALES"].sum() / \
                 data.groupby(["LIFESTAGE", "PREMIUM_CUSTOMER"])["PROD_QTY"].sum()
price_per_unit = price_per_unit.reset_index(name="Avg Price per Unit")
```

```python
# Combine all segment metrics into a summary table
segment_summary = pd.merge(sales_by_segment, customers_by_segment, on=["LIFESTAGE", "PREMIUM_CUSTOMER"])
segment_summary = pd.merge(segment_summary, units_per_customer, on=["LIFESTAGE", "PREMIUM_CUSTOMER"])
segment_summary = pd.merge(segment_summary, price_per_unit, on=["LIFESTAGE", "PREMIUM_CUSTOMER"])
```

```python
print("\n### Customer Segment Summary")
print(segment_summary.sort_values(by="Total Sales", ascending=False))
```

```python
# Brand Preference Analysis for Mainstream Young Singles/Couples
target_segment_data = data[(data["LIFESTAGE"] == "YOUNG SINGLES/COUPLES") & (data["PREMIUM_CUSTOMER"] == "Mainstream")]
other_segments_data = data[~((data["LIFESTAGE"] == "YOUNG SINGLES/COUPLES") & (data["PREMIUM_CUSTOMER"] == "Mainstream"))]
```

```python
target_brand_prop = target_segment_data.groupby("BRAND")["PROD_QTY"].sum() / target_segment_data["PROD_QTY"].sum()
other_brand_prop = other_segments_data.groupby("BRAND")["PROD_QTY"].sum() / other_segments_data["PROD_QTY"].sum()
brand_affinity = (target_brand_prop / other_brand_prop).sort_values(ascending=False).fillna(0)
```

```python
print("\n### Brand Affinity for Mainstream Young Singles/Couples (relative to other segments)")
print(brand_affinity.head(10))
```

```python
# Pack Size Preference Analysis for Mainstream Young Singles/Couples
target_pack_prop = target_segment_data.groupby("PACK_SIZE")["PROD_QTY"].sum() / target_segment_data["PROD_QTY"].sum()
other_pack_prop = other_segments_data.groupby("PACK_SIZE")["PROD_QTY"].sum() / other_segments_data["PROD_QTY"].sum()
pack_affinity = (target_pack_prop / other_pack_prop).sort_values(ascending=False).fillna(0)
```

```python
print("\n### Pack Size Affinity for Mainstream Young Singles/Couples (relative to other segments)")
print(pack_affinity.head(10))
```
